```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
Y = iris.target
```

```python
# Elbow curve to find the optimal number of clusters

wcss = []
for i in range(1, 11):
    KM = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    KM.fit(X)
    wcss.append(KM.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='blue')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS / Inertia')
plt.grid(True)
plt.show()
```

```python
optimal_k = 3
KM = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_KM = KM.fit_predict(X)
print("First 10 cluster assignments:", y_KM[:10])
print("\nFinal Cluster Centroids (3 centroids, 4 features each):\n", KM.cluster_centers_)
```
