```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
import pandas as pd
```

```python
df = load_iris()
X = df.data
Y = df.target

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

LR = LogisticRegression()
LRF = LR.fit(x_train,y_train)
y_pred = LRF.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:", accuracy*100)
print("Score:", LR.score(x_train,y_train)*100)
print("Coffecient",LR.coef_)
print("Intercept",LR.intercept_)
```

```python
MLR = LogisticRegression(multi_class='auto')
MLR.fit(x_train,y_train)
y_pred_multinomial = MLR.predict(x_test)
accuracy_multinomial = accuracy_score(y_test,y_pred_multinomial)
print("Accuracy:", accuracy_multinomial*100)
print("Score:", MLR.score(x_train,y_train)*100)
print("Coffecient",LR.coef_)
print("Intercept",LR.intercept_)
```

```python
L1Regularization = LogisticRegression(penalty='l1',solver='liblinear')
L2Regularization = LogisticRegression(penalty='l2',solver='lbfgs')
model = LogisticRegression(penalty="elasticnet",l1_ratio=0.5,solver='saga')

L1Regularization.fit(x_train,y_train)
y_pred_L1 = L1Regularization.predict(x_test)
accuracy_L1 = accuracy_score(y_test,y_pred_L1)
print("Accuracy L1:", accuracy_L1*100)
print("Score L1:", L1Regularization.score(x_train,y_train)*100)
print("Coffecient L1:",L1Regularization.coef_)
print("Intercept L1:",L1Regularization.intercept_)

L2Regularization.fit(x_train,y_train)
y_pred_L2 = L2Regularization.predict(x_test)
accuracy_L2 = accuracy_score(y_test,y_pred_L2)
print("Accuracy L2:", accuracy_L2*100)
print("Score L2:", L2Regularization.score(x_train,y_train)*100)
print("Coffecient L2:",L2Regularization.coef_)
print("Intercept L2:",L2Regularization.intercept_)

model.fit(x_train,y_train)
y_pred_mix = model.predict(x_test)
accuracy_mix = accuracy_score(y_test,y_pred_mix)
print("Accuracy mix:", accuracy_mix*100)
print("Score mix:", model.score(x_train,y_train)*100)
print("Coffecient mix:",model.coef_)
print("Intercept mix:",model.intercept_)
```
